```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Replace this with your actual data)
data = {
    'Platform': ['PS4', 'Xbox One', 'PS4', 'Xbox One', 'PC'],
    'Global_Sales': [80, 75, 50, 45, 30],
    'Video_Game': ['Game1', 'Game2', 'Game3', 'Game4', 'Game5'],
    'Year': [2019, 2019, 2020, 2020, 2021],
    'Country': ['EU', 'NA', 'JP', 'NA', 'EU'],
    'Genre': ['Action', 'Sports', 'Adventure', 'Sports', 'FPS'],
    'Publisher': ['Publisher1', 'Publisher2', 'Publisher1', 'Publisher2', 'Publisher3']
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'data' with your actual dataset
# Ensure you have columns like 'Platform', 'Global_Sales', etc.
sns.scatterplot(data=data, x='Platform', y='Global_Sales', size='Global_Sales', sizes=(20, 200))
plt.title('Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
top_10_games = df.groupby('Video_Game')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_games.plot(kind='bar', color='skyblue')
plt.title('Top 10 Video Games by Global Sales')
plt.xlabel('Video Game')
plt.ylabel('Global Sales')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Global_Sales', hue='Country', data=df[df['Country'].isin(['EU', 'JP', 'NA'])])
plt.title('Yearly Sales Revenue by Country')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()
```

```python
top_10_genres = df.groupby('Genre')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_genres.plot(kind='bar', color='lightcoral')
plt.title('Total Sales by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.show()
```

```python
top_10_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_publishers.plot(kind='bar', color='gold')
plt.title('Top 10 Publishers')
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.show()
```
